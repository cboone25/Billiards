```python
def ngon(n):
    table = []
    tab_x = []
    tab_y = []
    table.append([1,0,np.cos((2 * math.pi)/n),np.sin((2 * math.pi)/n), -1*np.cos((n-2)/(2*n)),
                  round(np.sin((n-2)/(2*n)),4)])
    tab_x.append(table[0][0])
    tab_y.append(table[0][1])
    tab_x.append(table[0][2])
    tab_y.append(table[0][3])
    for i in range(1,n+1):
        new_walls = [np.cos((i * 2 * math.pi)/n),np.sin((i*2 * math.pi)/n)
                     , np.cos((i+1)*2*math.pi/n), np.sin((i+1)*2*math.pi/n)
                  ,(np.cos((i+1)*2*math.pi/n)-np.cos((i)*2*math.pi/n)),
                     np.sin((i+1)*2*math.pi/n)-np.sin((i)*2*math.pi/n)]
        table.append(new_walls)
        tab_x.append(new_walls[0])
        tab_y.append(new_walls[1])
        tab_x.append(new_walls[2])
        tab_y.append(new_walls[3])
    return table,tab_x,tab_y
```

```python

import matplotlib.pyplot as plt

table,tab_x,tab_y = ngon(5)
plt.plot(tab_x,tab_y)
plt.show()
```

```python
def draw_table(h,l,num):
    theta =360/num

    x_end = 0
    y_end = 0
    u_x_r =0
    u_y_r = 1
    a =[h,l,h,l,l]
    table = []
    table_x = []
    table_y = []
    for i in range(num):
        x_start = x_end
        y_start = y_end
        u_x = u_x_r
        u_y = u_y_r
        x_end = x_start + (a[i] * (u_x))
        y_end = y_start + (a[i] * (u_y))

        u_x_r = (u_x * (math.cos(math.radians(theta)))) + ( u_y * (math.sin(math.radians(theta))))
        u_y_r = (-1 * u_x * (math.sin(math.radians(theta)))) +  (u_y * (math.cos(math.radians(theta)))) 
    #     print(theta)
        table.append([int(x_start),int(y_start),int(x_end),int(y_end),u_x,u_y])
        table_x.append(x_start)
        table_y.append(y_start)
        table_x.append(x_end)
        table_y.append(y_end)
    return table, table_x, table_y
```

```python
table, table_x, table_y = draw_table(5,5,5)

plt.plot(table_x, table_y)
plt.show()
```

```python
def convert_normal(v_x,v_y):
    v_x_unit = (v_x/((v_x*v_x + v_y*v_y)**(1/2)))
    v_y_unit = (v_y/((v_x*v_x + v_y*v_y)**(1/2)))
    return v_x_unit, v_y_unit
```

```python
def cal_dist(table):
    wall_dist = []
    for i in table:
        x_diff = (i[2] - i[0]) ** 2
        y_diff = (i[3] - i[1]) ** 2
        dist = math.sqrt(x_diff+y_diff)
        wall_dist.append(dist)
    return wall_dist
```
